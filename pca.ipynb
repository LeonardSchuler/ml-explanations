```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#sns.set_theme(style="whitegrid")
```

```python
n = 100
```

```python
x = np.linspace(0, 10, n)
```

```python
eps = np.random.normal(size=n)
```

```python
y = x + eps
```

```python
X = np.vstack([x,y]).T
```

```python
U, S, VT = np.linalg.svd(X, full_matrices=False)
U = -1*U
VT = -1*VT
V = VT.T
```

```python
fig, ax = plt.subplots()

ax.set_xlim(-2, 12)
ax.set_ylim(-2, 12)
ax.set_xticks(range(11))
ax.set_yticks(range(11))
ax.arrow(0,0,0,10, width=0.01, color='black', head_width=0.2)
ax.arrow(0,0,10,0, width=0.01, color='black', head_width=0.2)

sns.scatterplot(x=x, y=y, color='black', ax=ax)


plt.axis('equal')
plt.show()
```
